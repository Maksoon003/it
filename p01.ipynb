```python
df = df[['id', 'description', 'key_skills', 'schedule_id', 'schedule_name',
    'experience_id', 'experience_name',
       'specializations','employer_id', 'employer_name',
       'employer_vacancies_url',
       'employer_industries', 'type_id',
       'type_name', 'response_url','salary_from',
       'salary_to', 'salary_currency', 'archived', 'name',
       'area_id', 'area_name', 'area_url',
       'published_at', 'address_city',
       'apply_alternate_url', 'code','department_name',
       'employment_id', 'employment_name']]
```

```python
df
```

```python
df['published_at_dt'] = pd.to_datetime(df['published_at'], format='%Y-%m-%d')
df['year'] = df['published_at_dt'].dt.year
df['month'] = df['published_at_dt'].dt.month
df['day'] = df['published_at_dt'].dt.weekday

print(df['year'].value_counts(dropna=False))
print()
print(df['month'].value_counts(dropna=False))
print()
print(df['day'].value_counts(dropna=False))
print()
```
