```python
cols = df.columns[:54] # 54 колонок
# определяем цвета 
# желтый - пропущенные данные, синий - не пропущенные
colours = ['#000099', '#ffff00'] 
sns.heatmap(df[cols].isnull(), cmap=sns.color_palette(colours))
```

Приведенная ниже карта демонстрирует паттерн пропущенных значений для  54 признаков набора
По горизонтальной оси расположены признаки, по вертикальной – количество записей/строк 
Желтый цвет соответствует пропускам данных

  

```python
num_rows = len(df.index)   # список признаков, у которых более 95% строк содержат одно и то же значение.
low_information_cols = [] 

for col in df.columns:
    cnts = df[col].value_counts(dropna=False)
    top_pct = (cnts/num_rows).iloc[0]
    
    if top_pct > 0.95:
        low_information_cols.append(col)
        print('{0}: {1:.5f}%'.format(col, top_pct*100))
        print(cnts)
        print()
```
